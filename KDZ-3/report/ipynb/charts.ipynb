```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
verticles = pd.read_csv('..\\..\\tables\\_verticles.csv', sep=';', index_col='Search')
edges = pd.read_csv('..\\..\\tables\\_edges.csv', sep=';', index_col='Search')
time = pd.read_csv('..\\..\\tables\\_time.csv', sep=';', index_col='Search')
```

```python
y_max = max(verticles.max().max(), edges.max().max(), time.max().max()) * 1.1
y_min = min(verticles.min().min(), edges.min().min(), time.min().min()) * 0.9
```

```python
def print_two_plots(first_df, second_df, time_df, label, scale):
    fig, axes = plt.subplots(nrows=1, ncols=2)

    axes[0].plot(first_df[0:1].T, time_df[0:1].T, label='Полный граф')
    axes[0].plot(first_df[1:2].T, time_df[1:2].T, label='Связный граф')
    axes[0].plot(first_df[2:3].T, time_df[2:3].T, label='Разреженный граф')
    
    axes[1].plot(second_df[0:1].T, time_df[0:1].T, label='Полный граф')
    axes[1].plot(second_df[1:2].T, time_df[1:2].T, label='Связный граф')
    axes[1].plot(second_df[2:3].T, time_df[2:3].T, label='Разреженный граф')

    if (scale):
        axes[0].set_ylim(y_min, y_max)
        axes[1].set_ylim(y_min, y_max)

    axes[0].set_position([0, 0, 0.8, 0.85])
    axes[1].set_position([1, 0, 0.8, 0.85])

    axes[0].set_title('Вершины')
    axes[0].set_xlabel('Количество вершин')
    axes[0].set_ylabel('Время, нс')

    axes[1].set_title('Ребра')
    axes[1].set_xlabel('Количество ребер')
    axes[1].set_ylabel('Время, нс')

    axes[0].legend()
    axes[1].legend()

    fig.suptitle(label, x=0.8)

    plt.show()
```

```python
time
```

```python
print_two_plots(verticles[0:3], edges[0:3], time[0:3], 'Алгоритм Дейкстры', scale=False)
```

```python
print_two_plots(verticles[3:6], edges[3:6], time[3:6], 'Алгоритм Флойда-Уоршелла', scale=False)
```

```python
print_two_plots(verticles[6:9], edges[6:9], time[6:9], 'Алгоритм Беллмана-Форда', scale=False)
```

```python
print_two_plots(verticles[9:12], edges[9:12], time[9:12], 'Алгоритм A*', scale=False)
```

```python
first_df = verticles
second_df = edges
time_df = time

scale = False

fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(first_df[0:1].T, time_df[0:1].T, label='Дейкстра')
axes[1].plot(second_df[0:1].T, time_df[0:1].T, label='Дейкстра')


axes[0].plot(first_df[3:4].T, time_df[3:4].T, label='Флойд-Уоршелл')
axes[1].plot(second_df[3:4].T, time_df[3:4].T, label='Флойд-Уоршелл')

axes[0].plot(first_df[6:7].T, time_df[6:7].T, label='Беллман-Форд')
axes[1].plot(second_df[6:7].T, time_df[6:7].T, label='Беллман-Форд')

axes[0].plot(first_df[9:10].T, time_df[9:10].T, label='A*')
axes[1].plot(second_df[9:10].T, time_df[9:10].T, label='A*')

if (scale):
    axes[0].set_ylim(y_min, y_max)
    axes[1].set_ylim(y_min, y_max)

axes[0].set_position([0, 0, 0.8, 0.85])
axes[1].set_position([1, 0, 0.8, 0.85])

axes[0].set_title('Вершины')
axes[0].set_xlabel('Количество вершин')
axes[0].set_ylabel('Время, нс')

axes[1].set_title('Ребра')
axes[1].set_xlabel('Количество ребер')
axes[1].set_ylabel('Время, нс')

axes[0].legend()
axes[1].legend()

fig.suptitle('Все алгоритмы, полный граф', x=0.8)

plt.show()
```
