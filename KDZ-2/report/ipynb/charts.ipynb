```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
clear_small = pd.read_csv('..\\..\\tables\\clear_small_time.csv', sep=';', index_col='Search')
clear_big = pd.read_csv('..\\..\\tables\\clear_big_time.csv', sep=';', index_col='Search')

mod_small = pd.read_csv('..\\..\\tables\\mod_small_time.csv', sep=';', index_col='Search')
mod_big = pd.read_csv('..\\..\\tables\\mod_big_time.csv', sep=';', index_col='Search')
```

```python
y_max = max(clear_small.max().max(), clear_big.max().max(), mod_small.max().max(), mod_big.max().max()) * 1.1
y_min = min(clear_small.min().min(), clear_big.min().min(), mod_small.min().min(), mod_big.min().min()) * 0.9
```

```python
def get_first_part_of_index_name_by_row(df, row_number):
    index_name = df.index[row_number]
    first_part_of_name = index_name.split()[0]
    return first_part_of_name
```

```python
def print_two_clear_plots(first_df, second_df, label, scale):
    fig, axes = plt.subplots(nrows=1, ncols=2)

    first_df.T.plot(ax=axes[0], legend=False)
    second_df.T.plot(ax=axes[1], legend=False)

    if (scale):
        axes[0].set_ylim(y_min, y_max)
        axes[1].set_ylim(y_min, y_max)

    axes[0].set_position([0, 0, 0.8, 0.85])
    axes[1].set_position([1, 0, 0.8, 0.85])

    axes[0].set_title('Размер строки 10000')
    axes[0].set_xlabel('Длина шаблона поиска')
    axes[0].set_ylabel('Время, нс')

    axes[1].set_title('Размер строки 100000')
    axes[1].set_xlabel('Длина шаблона поиска')
    axes[1].set_ylabel('Время, нс')

    fig.suptitle(label, x=0.8)

    plt.show()
```

```python
def print_two_mod_plots(first_df, second_df, label, scale):
    fig, axes = plt.subplots(nrows=1, ncols=2)

    first_df.T.plot(ax=axes[0], legend=False)
    second_df.T.plot(ax=axes[1], legend=False)

    if (scale):
        axes[0].set_ylim(y_min, y_max)
        axes[1].set_ylim(y_min, y_max)

    axes[0].set_position([0, 0, 0.8, 0.85])
    axes[1].set_position([1, 0, 0.8, 0.85])

    axes[0].legend(['1 символ подстановки', '2 символа подстановки', '3 символа подстановки', '4 символа подстановки'],
        loc=(0.5, 1.1), ncol=2)

    axes[0].set_title('Размер строки 10000')
    axes[0].set_xlabel('Длина шаблона поиска')
    axes[0].set_ylabel('Время, нс')

    axes[1].set_title('Размер строки 100000')
    axes[1].set_xlabel('Длина шаблона поиска')
    axes[1].set_ylabel('Время, нс')

    fig.suptitle(label, x=0.8, y=1.1)

    plt.show()
```

```python
print_two_clear_plots(clear_small[0:1], clear_big[0:1], 'Наивный алгоритм с посимвольным сравнением', scale=True)
```

```python
print_two_clear_plots(clear_small[0:1], clear_big[0:1], 'Наивный алгоритм с посимвольным сравнением', scale=False)
```

```python
print_two_clear_plots(clear_small[1:2], clear_big[1:2], 'КМП, стандартные грани с посимвольным сравнением', scale=True)
```

```python
print_two_clear_plots(clear_small[2:3], clear_big[2:3], 'КМП, уточненные грани с посимвольным сравнением', scale=True)
```

```python
print_two_mod_plots(mod_small[0:4], mod_big[0:4], 'Наивный алгоритм с посимвольным сравнением, 1-4 символов подстановки', scale=True)
```

```python
print_two_mod_plots(mod_small[0:4], mod_big[0:4], 'Наивный алгоритм с посимвольным сравнением, 1-4 символов подстановки', scale=False)
```
